```
# the number of training samples 
num_train = 8000
# as a rule of thumb, 70-90% goes to training and remaining goes to validation
valid_size = .2
# performing operation on indices so, we can split array with same images and labels
indices = list(range(num_train))
# split the dataset
split = int(np.floor(valid_size*num_train))
# randomly generate it so, we don't have unbalanced class
np.random.shuffle(indices)
# now dividing it into training and test indices:
train_idx, test_idx = indices[split:], indices[:split]
# apply the operation on dataset:
# validation
X_val = images[test_idx]
y_val = labels[test_idx]
# training:
X_train = images[train_idx]
y_train = labels[train_idx]
```



Now, comes the deep learning technique we will using to train our classifier. 

I have used KNN i.e., ***K - nearest neighbor*** because of its*** easier to implement*** and ***simple to understand*** how image classification works on greyscale dataset.

***The Nearest neighbor is basic version of KNN, it compare each every image from test to train set and find the one that is closest for every image in test set.***

To do that we need a criteria to compare images; there are various ***distance matrices*** available; I have implemented ***L1 distance*** and ***L2*** to check which one give higher accuracy

```
# Nearest Neighbor classifier
class NearestNeighbor(object):
  # initialize the classifier
  def __init__(self):
    pass
  # train the classifier
  def train(self, X, y):
    """ X is N x D where each row is an example. Y is 1-dimension of size N """
    # the nearest neighbor classifier simply remembers all the training data
    self.Xtr = X
    self.ytr = y
  # prediction function for new data
  def predict(self, X):
    """ X is N x D where each row is an example we wish to predict label for """
    num_test = X.shape[0]
    # lets make sure that the output type matches the input type
    Ypred = np.zeros(num_test, dtype = self.ytr.dtype)

    # loop over all test rows
    for i in range(num_test):
      # find the nearest training image to the i'th test image
      # using the L1 distance (sum of absolute value differences)
      distances = np.sum(np.abs(self.Xtr - X[i,:]), axis = 1)
      min_index = np.argmin(distances) # get the index with smallest distance
      Ypred[i] = self.ytr[min_index] # predict the label of the nearest example

    return Ypred
```

The nearest neighbor classifier works on distance matrix and to compute distance between images we need to flatten all our pixels in a single row

```
# Preprocessing again:
# reshape the image data into rows
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_val = np.reshape(X_val, (X_val.shape[0], -1))

# As a sanity check, print out the shapes of the data
print('Training data shape: ', X_train.shape)
print('Validation data shape: ', X_val.shape)
```

```
# using L1 distance
nn = NearestNeighbor() # create a Nearest Neighbor classifier class
nn.train(X_train, y_train) # train the classifier on the training images and labels
Yte_predict = nn.predict(X_val) # predict labels on the test images
# and now print the classification accuracy, which is the average number
# of examples that are correctly predicted (i.e. label matches)
print ('accuracy: %f' % ( np.mean(Yte_predict == y_val) ))
```

```
#using L2 distance matrix:
class NearestNeighbor(object):
  # intialize the classifier
  def __init__(self):
    pass
  # train classifier on dataset
  def train(self, X, y):
    """ X is N x D where each row is an example. Y is 1-dimension of size N """
    # the nearest neighbor classifier simply remembers all the training data
    self.Xtr = X
    self.ytr = y
  # to make prediction on test set
  def predict(self, X):
    """ X is N x D where each row is an example we wish to predict label for """
    num_test = X.shape[0]
    # lets make sure that the output type matches the input type
    Ypred = np.zeros(num_test, dtype = self.ytr.dtype)

    # loop over all test rows
    for i in range(num_test):
      # find the nearest training image to the i'th test image
      # using the L2 distance (square root of sum of square value differences)
      distances = np.sqrt(np.sum(np.square(self.Xtr - X[i,:]), axis = 1))
      # distances = np.sum(np.abs(self.Xtr - X[i,:]), axis = 1)
      # nearest neighbor:
      min_index = np.argmin(distances) # get the index with smallest distance
      Ypred[i] = self.ytr[min_index] # predict the label of the nearest example

    return Ypred
```

```
# now we will be testing same approach but with finding k closest images
# and then making prediction on them
class KNearestNeighbor(object):
    """ a kNN classifier with L2 distance """

    def __init__(self):
        self.x_train = None
        self.y_train = None

    def train(self, x, y):
        """
        Train the classifier. For k-nearest neighbors this is just
        memorizing the training data.
        :param x: A numpy array of shape (num_train, D) containing the training data
                  consisting of num_train samples each of dimension D.
        :param y: A numpy array of shape (N,) containing the training labels, where
                  y[i] is the label for X[i].
        :return: None
        """

        self.x_train = x
        self.y_train = y
    
    def predict(self, x, k=1):
        """
        Predict labels for test data using this classifier.
        :param x: A numpy array of shape (num_test, D) containing test data consisting
                  of num_test samples each of dimension D.
        :param k: The number of nearest neighbors that vote for the predicted labels.

        :return y: A numpy array of shape (num_test,) containing predicted labels for the
                   test data, where y[i] is the predicted label for the test point X[i].
        """
        
        dists = self.compute_distances(x)

        return self.predict_labels(dists, k=k)

    def compute_distances(self, x):
        """
        Compute the distance between each test point in X and each training point
        in self.x_train.
        :param x: A numpy array of shape (num_test, D) containing test data.
        :return dists: A numpy array of shape (num_test, num_train) where dists[i, j]
                      is the Euclidean distance between the ith test point and the
                      jth training point.
        """

        # shape (num_test, 1)
        x_norms = np.sum(x ** 2, axis=1).reshape(-1, 1)
        # shape (num_train, )
        x_train_norms = np.sum(self.x_train ** 2, axis=1)
        # summing arrays of shapes (num_test, 1) and (num_train, )
        # gives exactly the necessary shape (num_test, num_train). 
        # It's broadcasting, a technique to manipulate matrices of not equal sizes
        dists = (x_norms + x_train_norms - 2 * x.dot(self.x_train.T)) ** .5

        return dists

    def predict_labels(self, dists, k=1):
        """
        Given a matrix of distances between test points and training points,
        predict a label for each test point.
        :param dists: A numpy array of shape (num_test, num_train) where dists[i, j]
                      gives the distance between the ith test point and the jth training point.
        :param k: number of neighbors in kNN (hyper-parameter)
        :return y: A numpy array of shape (num_test,) containing predicted labels for the
                   test data, where y[i] is the predicted label for the test point X[i].
        """
        num_test = dists.shape[0]
        y_pred = np.zeros(num_test)
        for i in range(num_test):
            # finding the nearest neighbor's indices from distance matrix
            closest_idx = np.argsort(dists[i, :])[:k]
            # determining the labels
            neighbor_labels = self.y_train[closest_idx]
            # ravel() for easier vector computation
            neighbor_labels = neighbor_labels.ravel()
            # counting the most occured label based on nearest neighbor
            label_counts = np.bincount(neighbor_labels)
            # predicting the value of maximum occurence.
            y_pred[i] = np.argmax(label_counts)

        return y_pred
```

Let us test with all k values and see which one gives maximum accuracy

To do that, we will be tunig a hyperparameter known as K. The process of arbitrarily picking the first 1000 datapoints to be validation set and remaining one as training set is known as ***cross-validation***.  

```
# to calculate how much time it takes for classifier to run
%%time
# import tqdm for display of progress box - Hbox
from tqdm import tqdm_notebook

# dividing training sets in five folds and combining them randomly to 
# determine highest accuracy
num_folds = 5
# values of nearest neighbor
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = []
y_train_folds = []
# split the training set indices using numpy.array_split function
split_idx = np.array_split(range(X_train.shape[0]), num_folds)
# combining folds randomly
X_train_folds = np.array([X_train[train_idx] for train_idx in split_idx])
y_train_folds = np.array([y_train[train_idx].reshape(-1,1)
                          for train_idx in split_idx])
# creating a dictionary to keep track of accuracy from different folds
k_to_accuracies = {}

# intialize the nearest neighbor
knn = KNearestNeighbor()
# loop around every choice of nearest neighbor 
for k in tqdm_notebook(k_choices):
  # for each value calculate accuracy
    accuracies = []
    # for each value of a fold in training set
    for i in range(num_folds):
      # join training set as one and keeping validation set as another
        all_other_fold_idx = list(set(range(num_folds)).difference([i]))
        X_train_cv = np.vstack(X_train_folds[all_other_fold_idx])
        y_train_cv = np.vstack(y_train_folds[all_other_fold_idx]).flatten()
        X_valid_cv = X_train_folds[i]  
        y_valid_cv = y_train_folds[i].flatten()
        # training classifier using training set
        knn.train(X_train_cv, y_train_cv)   
        # predicting classifier using validation set
        knn_valid_pred = knn.predict(X_valid_cv, k=k)

        # calculating accuracy of each fold
        num_correct = np.sum(knn_valid_pred == y_valid_cv)
        # determing number of correct values of labels
        accuracy = float(num_correct) / y_valid_cv.shape[0]
        # joining them into list
        accuracies.append(accuracy)
        # making a dictionary of every folds accuracy
    k_to_accuracies[k] = accuracies

 # Print out the computed accuracies
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
```

Let us take a look at accuracy using cross-validation and plot the values of k for better understanding of classifier we have train

```
# plot the raw observations
for k in k_choices:
    accuracies = k_to_accuracies[k]
    plt.scatter([k] * len(accuracies), accuracies)

# plot the trend line with error bars that correspond to standard deviation
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
```

As we can see maximum accuracy is between k = 5 to k= 15
